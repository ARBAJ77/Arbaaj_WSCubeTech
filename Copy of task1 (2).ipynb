```python
#find area of a trapezoid
short_length = float(input('enter shorter length: '))
longer_length = float(input('enter longer length: '))
height = float(input('enter height: '))

area = ((short_length+longer_length)/2*height)
print(area)
```

```python
# Write a program to identify if the number given by the user is a 1 digit number, 2 digit number and so on upto 5.

num = int(input("enter a number upto 5 digit: "))

if num >=0 and num<=9:
   print("1 digit")
elif num >=10 and num <= 99:
   print("2 digit")
elif num >=100 and num <=999:
   print("3 digit")
elif num >=1000 and num <=9999:
    print("4 digit") 
elif num >10000 and num <=99999:
    print("5 digit")
    

```
