```python
#write a program to find a factorial of a number
# factorial of 0 is 1
num = int(input("enter a number here :"))
fact= 1

for i in range(1,num+1):
    fact =fact*i
print("factorial of", num,"is",fact)    
```

```python
#find the sum of all the even numbers upto 50

sum = 0
for i in range(0,51):  
     if i % 2 == 0 :
       sum+=i
       print(sum)
```

```python
#find the sum of all the odd numbers upto 70

sum = 0
for i in range(0,70):
     if i % 2 != 0:
        sum+= i
        print(sum)
```
